```python




# creatting function n applying into any column to do some modification 

def increment_age_by_5(x):
    return x+100
obesityDataSet['Age'] = obesityDataSet['Age'].apply(increment_age_by_5)

obesityDataSet['Age']
```

```python
# subtracting age by 5 
def subtractingAgeBy5(x):
    print(x,end=" . ")
    return x-10

obesityDataSet['Age'] = obesityDataSet['Age'].apply(subtractingAgeBy5)
obesityDataSet['Age']
```
